## **"Print statement"** 

The print () function allows to display a message to the standard output (screen) or in the file of the same text. An example that can be presented is that of ("WELCOME") and that word will appear in the program.

```python
print("Hi guys", "how are you today?")
```

### **Characteristics of Print Statement**


---
1) **Objects**  Presents that there may be more than one message to be printed.

2) **Sep** -The words are separated by quotation marks. ***Example:*** ('' ")

3) **End**  It is used to mark that a message will continue to be written.

4) **File**  They are multiple words with writing methods (string). If omitted, (sys.stdout) will be used, which prints messages to the screen.


### **Printing one or more variables**


---
For python to show us the value of a variable it is necessary to give the variable a name and value, then a print is placed at the beginning to present the value as a result.

***Example:***

```python
x = 2
print("In", x , " weeks there are ", 7 * x, "days.")
```

### **Quotation marks in quotation marks**


---

***Example:***

```python
print("A boy asks his friend : \"How were you yesterday?\"")
print(' And the other boy answer: \' I was excellent!\'')
```

## **"If/ else statement"**

In programming, If/else statement is necessary when we want to execute a code alone or if we want to fulfill a certain condition. That is what if / else conditionals are used for, and, each one has a different use.

Imagen 8.jpg

### **If statement concept**


---

In Python, the if conditional is indicated by a different color in the program. The data begins with an if and the data that can be true or false. 

As an additional data, in python a number that is greater than zero means that the data is true, but if it is equal to or less than zero, it means that it is false.


### **Characteristics of the if conditional**


---


1) The if conditional statement is used to make decisions, it can also basically evaluate a logical operation.

2) To express that if a quantity expresses a different result if it is greater or less, the conditional (if) and some other symbols must be known to give that expected result. ***For example: (>, <).***


Imagen 9.JPG

**Example 1: Know if a number is greater than 0**



```python
value = 8
if value > 0:
    print(value, "The number entered is greater than 0.")
```

**Example 2: Price comparator**

```python
money = int(input("How much money do you have to spend?"))
if money < 15:
    print("You can have a soda or some chips")
if money > 25:
    print("You can choose what you want!")
```

### **Else statement concept**


---
The else command tells Python that the block that follows must be executed when the previous condition has not been fulfilled (that is, when it is false), this condition must by a colon (:).



### **Characteristics of the else conditional**


---

1) Using the else conditional we save ourselves from writing a condition and avoid errors in the program.

2) The else conditional will be executed only if none of the above conditions are met.

3) If different if / else conditions are put together, a new condition can be created, called ***else if.***

Imagen 10.JPG

**Example 1: Exam check**

```python
points = int(input("How many points did you get on the exam? "))
if points < 80:
    print("You will have to take my exam again")
else:
    print("Congratulations, you passed my exam!")
```

**Example 2: Trip to Disney**

```python
calif = int(input("How much did you get out of the qualification? "))
if calif >= 8:
    print("OK, you can go to Disney with us")
else:
    print("You will stay to study all vacations!")
```

## **"Loops"**

A loop is the number of times a data is repeated, until it is complete. A loop usually has an initial data, a data sequence and finally a limit data that ends the loop. The two loops in the programming that are the best known are: **the for loop and the while loop.**

### **Characteristics of the Loops**


---

1) Loops help to reduce the number of times a code that manages to execute in the same operation multiple times is repeated.

2) To loop through a set of code a specified number of times, we can use the range () function.

3) It is important to understand the different types of loops that can be used, and to know that you can use the else conditional and other examples.


### **For loop**


---
The for loops allow us to execute one or more lines of repetitive code but having control and knowledge of the values. It is necessary to have an initial value and a final value.


Imagen 11.JPG

### **While loop**


---

A while loop allows repeating the execution of a certain group of instructions, if one or more conditions are met, if the condition is not met, the cycle will stop. After the rest of the loop runs, the process repeats over and over if the condition is true, until the limit value is reached.

Imagen 12.JPG

**Example 1: For loop, sum from 1 to 100**

```python
for x in range(0, 101, +1):
  print(x)
```

**Example 2: Hours of study**

```python
data = int(input("Write a number less than 50: "))
while data >= 50:
    print("You have exceeded the limit value! Try again")
    data = int(input("Write a number less than 50: "))
print("Accepted value!")
print("Restart the program")
```
